#plt.title(".....") ---> Titlul graficului 

```python
x = [10,10]
y = [8,9]
plt.title("Acesta este titlul meu interesant")
plt.scatter(x,y);
plt.show()
```

```python
x = [10,10]
y = [8,9]
plt.title("Titlu")
plt.xlabel("Valoarea x")
plt.scatter(x,y);
plt.show()
```

```python
x = [10,10]
y = [8,9]
plt.title("Titlu")
plt.xlabel("Valoarea x")
plt.ylabel("Valoarea y")
plt.scatter(x,y);
plt.show()
```

###SALVAREA FIGURII IN FOLDER

```python
x = [10,10]
y = [8,9]
plt.title("Titlu")
plt.xlabel("Valoarea x")
plt.ylabel("Valoarea y")
plt.scatter(x,y);
plt.show()

plt.savefig("Figura_mea.png")
```

```python
a = [10,10,11,20]
b = [8,9,10,10]
plt.title("Titlu")
plt.xlabel("Valoarea x")
plt.ylabel("Valoarea y")
plt.scatter(a,b);
plt.show()

plt.savefig("Figura_mea.png")
```
